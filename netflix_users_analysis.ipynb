Максимальное количчество клиентов (415) было привлечено в ноябре 2022. На втором месте август 2022. Минимальное количество (0) было в июле и августе 2021 года.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime

# Преобразуем даты в datetime формат
df['join_date'] = pd.to_datetime(df['join_date'])
df['last_payment_date'] = pd.to_datetime(df['last_payment_date'])

# Рассчитываем продолжительность подписки в месяцах для каждого пользователя
df['months_active'] = ((df['last_payment_date'] - df['join_date']).dt.days / 30).apply(np.ceil).clip(lower=1)
df['total_revenue'] = df['monthly_revenue'] * df['months_active']

# Создаем временной ряд с полным диапазоном дат
min_date = df['join_date'].min()
max_date = df['last_payment_date'].max()
date_range = pd.date_range(start=min_date, end=max_date, freq='M')

# Создаем DataFrame для хранения накопленной выручки
cumulative_revenue = pd.DataFrame(index=date_range, columns=['revenue'])
cumulative_revenue['revenue'] = 0

# Рассчитываем накопленную выручку по месяцам
for date in date_range:
    # Фильтруем пользователей, которые присоединились до текущей даты
    active_users = df[df['join_date'] <= date]

    # Рассчитываем количество месяцев активности для каждого пользователя на текущую дату
    active_users['months_to_date'] = ((date - active_users['join_date']).dt.days / 30).clip(upper=active_users['months_active'])

    # Рассчитываем выручку для каждого пользователя на текущую дату
    active_users['revenue_to_date'] = active_users['monthly_revenue'] * active_users['months_to_date']

    # Суммируем выручку всех пользователей
    cumulative_revenue.loc[date, 'revenue'] = active_users['revenue_to_date'].sum()

# Создаем график
plt.figure(figsize=(16, 10))

# Линейный график накопленной выручки
plt.plot(
    cumulative_revenue.index,
    cumulative_revenue['revenue'],
    color='#FFA500',  # оранжевый
    linewidth=4,
    marker='o',
    markersize=8,
    markerfacecolor='#FFD700',  # золотистый
    markeredgecolor='#8B7500',  # темно-золотой
    markeredgewidth=1.5
)

# Заполнение области под линией
plt.fill_between(
    cumulative_revenue.index,
    cumulative_revenue['revenue'],
    color='#FFFACD',  # светло-желтый
    alpha=0.4
)

# Добавляем подписи значений
for date, revenue in zip(cumulative_revenue.index, cumulative_revenue['revenue']):
    plt.text(
        date,
        revenue + (0.03 * cumulative_revenue['revenue'].max()),
        f'{revenue:,.0f}',
        ha='center',
        va='bottom',
        fontsize=10,
        fontweight='bold',
        color='#8B7500'  # темно-золотой
    )

# Настройка оформления
plt.title('Накопленная выручка сервиса по месяцам', fontsize=18, pad=20)
plt.xlabel('Дата', fontsize=14)
plt.ylabel('Накопленная выручка', fontsize=14)
plt.grid(True, alpha=0.2, linestyle='--')

# Форматирование осей
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: f'${x/1e6:,.1f}M'))
plt.xticks(rotation=45, fontsize=11)
plt.gca().set_facecolor('#FFFFE0')  # светло-желтый фон

# Установка диапазона дат
plt.xlim(min_date - pd.DateOffset(months=1), max_date + pd.DateOffset(months=1))
plt.ylim(0, cumulative_revenue['revenue'].max() * 1.15)

# Добавляем информационную аннотацию
total_revenue = df['total_revenue'].sum()
avg_monthly = df['monthly_revenue'].mean()
plt.annotate(
    f'Общая выручка: ${total_revenue:,.0f}\nСредний ежемесячный платеж: ${avg_monthly:,.2f}',
    xy=(0.98, 0.02),
    xycoords='axes fraction',
    ha='right',
    va='bottom',
    fontsize=12,
    bbox=dict(boxstyle='round,pad=0.5', fc='#FFF9C4', ec='#FFD700', alpha=0.8)
)

plt.tight_layout()
plt.show()
```

Выручка поступала в динамике от $27 до  $355,797. Средний ежемесячный платеж $12,51


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Создаем возрастные группы с шагом 5 лет
age_bins = range(15, 100, 5)
age_labels = [f'{i}-{i+4}' for i in age_bins[:-1]]
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels, right=False)

# Рассчитываем общую выручку для каждого пользователя
df['total_revenue'] = df['monthly_revenue'] * ((df['last_payment_date'] - df['join_date']).dt.days // 30 + 1)

# Группируем по возрастным группам и вычисляем метрики платежеспособности
age_group_stats = df.groupby('age_group').agg(
    total_revenue=('total_revenue', 'sum'),
    avg_monthly_revenue=('monthly_revenue', 'mean'),
    avg_total_revenue=('total_revenue', 'mean'),
    user_count=('user_id', 'count')
).reset_index()

# Определяем самую платежеспособную группу по средней общей выручке
most_paying_group = age_group_stats.loc[age_group_stats['avg_total_revenue'].idxmax()]

# Создаем цветовую палитру (желтые оттенки)
yellow_palette = plt.cm.autumn_r(np.linspace(0.2, 0.8, len(age_group_stats)))
yellow_palette = [tuple(c) for c in yellow_palette[:, :3]]  # Преобразуем в RGB

# Создаем график
plt.figure(figsize=(14, 8))

# Столбчатая диаграмма средней общей выручки на пользователя
bars = plt.bar(
    age_group_stats['age_group'],
    age_group_stats['avg_total_revenue'],
    color=yellow_palette,
    edgecolor='#333333',
    linewidth=1.2
)

# Выделяем самую платежеспособную группу
most_paying_index = age_group_stats['age_group'].tolist().index(most_paying_group['age_group'])
bars[most_paying_index].set_edgecolor('red')
bars[most_paying_index].set_linewidth(3)

# Добавляем подписи значений
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height + (0.02 * max(age_group_stats['avg_total_revenue'])),
        f'${height:,.0f}',
        ha='center',
        va='bottom',
        fontsize=10,
        fontweight='bold'
    )

# Добавляем аннотацию для самой платежеспособной группы
plt.annotate(
    f"Самая платежеспособная группа: {most_paying_group['age_group']}\n"
    f"Средняя общая выручка: ${most_paying_group['avg_total_revenue']:,.0f}\n"
    f"Пользователей: {most_paying_group['user_count']:,}",
    xy=(most_paying_index, most_paying_group['avg_total_revenue']),
    xytext=(0.5, 0.9),
    textcoords='axes fraction',
    ha='center',
    arrowprops=dict(facecolor='red', shrink=0.05),
    bbox=dict(boxstyle='round,pad=0.5', fc='#FFF9C4', ec='red', alpha=0.9),
    fontsize=12
)

# Настройка оформления
plt.title('Средняя общая выручка по возрастным группам', fontsize=16, pad=20)
plt.xlabel('Возрастная группа', fontsize=14)
plt.ylabel('Средняя общая выручка на пользователя ($)', fontsize=14)
plt.xticks(fontsize=11)
plt.grid(axis='y', alpha=0.2, linestyle='--')

# Добавляем дополнительную информацию
plt.figtext(
    0.5, 0.01,
    f"Всего пользователей: {df['user_id'].nunique():,} | Общая выручка: ${df['total_revenue'].sum():,.0f}",
    ha='center',
    fontsize=12,
    bbox=dict(boxstyle='round', facecolor='#FFF9C4', alpha=0.5)
)

plt.tight_layout()
plt.subplots_adjust(bottom=0.15)  # Место для подписи внизу
plt.show()

# Выводим таблицу с результатами
print("\nДетальная статистика по возрастным группам:")
print(age_group_stats.sort_values('avg_total_revenue', ascending=False))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Создаем возрастные группы с шагом 5 лет
age_bins = range(15, 100, 5)
age_labels = [f'{i}-{i+4}' for i in age_bins[:-1]]
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels, right=False)

# Группируем данные по возрастной группе и типу подписки
subscription_dist = df.groupby(['age_group', 'subscription_type']).size().unstack().fillna(0)

# Рассчитываем долю каждого типа подписки в группе
subscription_percent = subscription_dist.div(subscription_dist.sum(axis=1), axis=0) * 100

# Находим группу с максимальной долей премиум-подписки
premium_max_group = subscription_percent['Premium'].idxmax()
premium_max_value = subscription_percent['Premium'].max()

# Создаем желтую палитру с акцентом на премиум
palette = {
    'Basic': '#FFD700',    # золотистый
    'Standard': '#FFA500', # оранжевый
    'Premium': '#FF8C00'   # темно-оранжевый (акцент)
}

# Создаем фигуру с двумя графиками
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(16, 14), gridspec_kw={'height_ratios': [2, 1]})

# График 1: Распределение типов подписок по возрастным группам (столбчатая диаграмма)
subscription_percent.plot(kind='bar', stacked=True, ax=ax1, color=[palette[col] for col in subscription_percent.columns])
ax1.set_title('Распределение типов подписок по возрастным группам', fontsize=18, pad=20)
ax1.set_xlabel('Возрастная группа', fontsize=14)
ax1.set_ylabel('Доля подписок (%)', fontsize=14)
ax1.legend(title='Тип подписки', fontsize=12)
ax1.grid(axis='y', alpha=0.2, linestyle='--')

# Добавляем подписи для премиум-подписок
for i, (idx, row) in enumerate(subscription_percent.iterrows()):
    premium_pct = row['Premium']
    if premium_pct > 5:  # Показываем только значимые значения
        ax1.text(i, row['Basic'] + row['Standard'] + premium_pct/2,
                f'{premium_pct:.1f}%',
                ha='center', va='center', fontsize=10, fontweight='bold', color='white')

# Выделяем группу с максимальной долей премиум
max_idx = subscription_percent.index.get_loc(premium_max_group)
ax1.patches[max_idx * 3 + 2].set_edgecolor('red')
ax1.patches[max_idx * 3 + 2].set_linewidth(2)

# График 2: Средняя выручка по возрастным группам
revenue_by_age = df.groupby('age_group')['monthly_revenue'].mean()
ax2.bar(revenue_by_age.index, revenue_by_age.values, color='#FFD700', edgecolor='#333333')
ax2.set_title('Средняя ежемесячная выручка по возрастным группам', fontsize=18, pad=20)
ax2.set_xlabel('Возрастная группа', fontsize=14)
ax2.set_ylabel('Средняя выручка ($)', fontsize=14)
ax2.grid(axis='y', alpha=0.2, linestyle='--')

# Добавляем подписи значений
for i, value in enumerate(revenue_by_age.values):
    ax2.text(i, value + 0.5, f'${value:.2f}', ha='center', va='bottom', fontsize=10)

# Выделяем группу с максимальной премиум-долей
ax2.bar(max_idx, revenue_by_age.iloc[max_idx], color='#FF8C00', edgecolor='red', linewidth=2)

# Добавляем аннотации
ax1.annotate(f"Максимальная доля Premium: {premium_max_value:.1f}%\nв группе {premium_max_group}",
            xy=(max_idx, subscription_percent.loc[premium_max_group, 'Premium']),
            xytext=(0.7, 0.95),
            textcoords='axes fraction',
            arrowprops=dict(facecolor='red', shrink=0.05),
            bbox=dict(boxstyle='round,pad=0.5', fc='#FFF9C4', ec='red', alpha=0.9),
            fontsize=12)

# Тепловая карта корреляции
plt.figure(figsize=(10, 8))
corr_matrix = pd.crosstab(df['age_group'], df['subscription_type'], normalize='index')
sns.heatmap(corr_matrix, cmap='YlOrBr', annot=True, fmt='.2f', linewidths=0.5)
plt.title('Корреляция возрастных групп и типов подписки', fontsize=16)
plt.xlabel('Тип подписки', fontsize=12)
plt.ylabel('Возрастная группа', fontsize=12)
plt.tight_layout()

# Общая информация
plt.figure(figsize=(8, 4))
plt.figtext(0.5, 0.95,
           f"Анализ зависимости типа подписки от возраста\n"
           f"Самая платежеспособная группа: {premium_max_group} (Premium: {premium_max_value:.1f}%, Средняя выручка: ${revenue_by_age.loc[premium_max_group]:.2f})",
           ha='center', fontsize=14, bbox=dict(boxstyle='round', facecolor='#FFF9C4', alpha=0.7))

plt.tight_layout()
plt.subplots_adjust(top=0.85)
plt.show()

# Вывод таблицы с данными
print("\nТоп-5 возрастных групп по доле премиум-подписок:")
print(subscription_percent[['Premium']].sort_values('Premium', ascending=False).head(5))
```
